```python
import pandas as pd
import numpy as np
```

```python
financials = pd.read_csv('../data/financials.csv')
financials.head(11)
```

```python
financials = financials.sort_values(by='id', ascending=False)
financials.head(10)
```

```python
financials.shape
```

```python
# check what companies did not get scraped
companies = pd.read_csv('../data/company_names_400_pages.csv')
ids = companies['company_no']
```

```python
missing_ids = []
l = len(financials['id'].unique())
for num in ids:
    count = 0
    for num2 in financials['id'].unique():
        if num == str(num2):
            break
        else:    
            count += 1
    if count == l:
        missing_ids.append(num)
len(missing_ids)
```
