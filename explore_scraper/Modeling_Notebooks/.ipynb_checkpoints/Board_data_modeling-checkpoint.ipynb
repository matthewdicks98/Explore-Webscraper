## Creating train and test sets making sure that the balance in both remains the same

```python
from sklearn.model_selection import train_test_split
```

```python
# drop nans 
board_features.dropna(inplace=True)
# just use active assignments star
board_features.drop('total_active_assignments',axis = 1, inplace = True)
```

```python
# splitting the dataset into defaulters and non-defaulters to be train, test split evenly
defaults = board_features[board_features['default']==1]
non_defaults = board_features[board_features['default']==0]

# X and y for these two groups
X_default = defaults.drop('default',axis = 1)
y_default = defaults['default']

X_non_default = non_defaults.drop('default',axis = 1)
y_non_default = non_defaults['default']

# train, test split these two groups
X_train_default, X_test_default, y_train_default, y_test_default = train_test_split(X_default, y_default, 
                                                                                    test_size=0.3, 
                                                                                    random_state=42)

X_train_non_default, X_test_non_default, y_train_non_default, y_test_non_default = train_test_split(X_non_default, 
                                                                                                    y_non_default, 
                                                                                                    test_size=0.3,
                                                                                                random_state=42)

# adding these back together
X_train = X_train_default.append(X_train_non_default,ignore_index = True)
y_train = y_train_default.append(y_train_non_default, ignore_index=True)
X_test = X_test_default.append(X_test_non_default,ignore_index = True)
y_test = y_test_default.append(y_test_non_default, ignore_index=True)
```

## Single factor logistic regression (same method as previously used, except using Logit)

```python
X_train.info()
```

```python
p_values = []
auc_train = []
auc_test = []
for c in X_train.columns[2:]:
    
    X_train_2 = sm.add_constant(X_train[c])
    logit_model=sm.Logit(y_train, X_train_2)
    result=logit_model.fit()
    p_values.append(result.pvalues[1])

    y_pred_train = list(map(round,result.predict(X_train_2)))
    y_pred_test = list(map(round,result.predict(sm.add_constant(X_test[c]))))
    auc_train.append(roc_auc_score(y_train,y_pred_train))
    auc_test.append(roc_auc_score(y_test,y_pred_test)) 
```

```python
significance = pd.DataFrame(data=p_values, columns=['p-values'], index=X_train.columns[2:]).sort_values(by='p-values')
significance['significant'] = significance['p-values'].apply(lambda x: 'Yes' if x < 0.05 else 'No')
significance['auc_train'] = auc_train
significance['auc_test'] = auc_test

significance
```

Auc means that none of the models were able to correctly classify bankrupcy

## Build a simple logistic regression model to predict default rates

```python
X_train_2 = sm.add_constant(X_train.drop(['company','id'],axis=1))
logit_model = sm.Logit(y_train, X_train_2)
result = logit_model.fit()
```

### Feature importance (p-values and confidense intervals)

```python
print(result.summary())
```

### Performance

```python
y_pred_train = list(map(round,result.predict(X_train_2)))
y_pred_test = list(map(round,result.predict(sm.add_constant(X_test.drop(['id','company'],axis=1)))))
print(roc_auc_score(y_train,y_pred_train))
print(roc_auc_score(y_test,y_pred_test)) 
```

```python
print("Train")
print(confusion_matrix(y_train,y_pred_train))
print("Test")
print(confusion_matrix(y_test,y_pred_test))
```

## Build a rf model

```python
rf_def = RandomForestClassifier(n_jobs=-1)
```

```python
rf_def.fit(X_train.drop(['company','id'],axis = 1), y_train)
```

```python
print(f"Confusion matrix = {confusion_matrix(y_train, y_pred_train)}")
print(f"Confusion matrix = {confusion_matrix(y_test, y_pred_test)}")
```

The random forest also did not learn much and it overfit (not surprisingly)
